# Ques.1 Write a program in python to visualize a user given data.

```python
import matplotlib.pyplot as plt
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie", "Diana", "Eve"],
    "Age": [25, 30, 28, 22, 32],
    "City": ["New York", "Los Angeles", "Chicago", "Seattle", "San Francisco"]
}

df = pd.DataFrame(data)
print(df)

plt.bar(df["Name"], df["Age"])  

plt.xlabel("Name")
plt.ylabel("Age")
plt.title("Age Distribution")
plt.xticks(rotation=45) 
plt.tight_layout()

plt.show()
```

# Ques.2 Write a program in python to analyze and perform the user data(create pie,scatter and bar chart). 

```python
import matplotlib.pyplot as plt
import pandas as pd


data = {
    "Name": ["Alice", "Bob", "Charlie", "Diana", "Eve"],
    "Age": [25, 30, 28, 22, 32],
    "City": ["New York", "Los Angeles", "Chicago", "Seattle", "San Francisco"]
}



df = pd.DataFrame(data)

# Create pie chart
plt.figure(figsize=(6, 6))  
plt.pie(df["Age"], labels=df["Name"], autopct="%1.1f%%")
plt.title("Pie Chart of Ages")
plt.show()

# Create scatter chart
plt.figure(figsize=(8, 6))
plt.scatter(df["Name"], df["Age"])
plt.xlabel("Name")
plt.ylabel("Age")
plt.title("Scatter Plot of Ages")
plt.grid(True)
plt.show()

# Create bar chart
plt.bar(df["Name"], df["Age"])
plt.xlabel("Name")
plt.ylabel("Age")
plt.title("Bar Chart of Ages")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Ques.3 Write a program in python to visualize the linear search of vector.  

```python
import matplotlib.animation as animation
import matplotlib.pyplot as plt


vector = [10, 20, 5, 30, 15]
target = 20  

# Initial plot
fig, ax = plt.subplots()
bars = ax.bar(range(len(vector)), vector, color='skyblue')
text = ax.text(0.5, 1.1, f"Target: {target}", ha="center", va="center", fontsize=14, transform=ax.transAxes)

# Animate the search process
def animate(i):
    if i == vector.index(target):
        bars[i].set_color('green')  
        text.set_text(f"Target Found at Index: {i}")
    else:
        bars[i].set_color('red')  

    return bars, text

anim = animation.FuncAnimation(fig, animate, frames=len(vector), interval=500)

plt.title("Linear Search Visualization")
plt.xlabel("Index")
plt.ylabel("Value")
plt.show()
```

# Ques.4 Write a program in python to visualize the Chi-square & poisson distribution.

# 1.Chi-Squared Distribution Visualization:

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import chi2


df = 5  
x = np.linspace(0.1, 10, 100)  
# probability density function (PDF) for the chi-squared distribution
pdf = chi2.pdf(x, df)

# plot
plt.figure(figsize=(8, 6))
plt.plot(x, pdf, label=f'Chi-Squared Distribution (df = {df})')

# labels, title, and legend
plt.xlabel('x')
plt.ylabel('PDF')
plt.title('Chi-Squared Distribution Visualization')
plt.legend()

# Show the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

# 2.Poisson Distribution Visualization:

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import poisson


lam = 5  
x = np.arange(0, 10)  

#  probability mass function (PMF) for the Poisson distribution
pmf = poisson.pmf(x, lam)

# bar chart
plt.figure(figsize=(8, 6))
plt.bar(x, pmf, label=f'Poisson Distribution ( = {lam})')

# labels, title, and legend
plt.xlabel('x')
plt.ylabel('PMF')
plt.title('Poisson Distribution Visualization')
plt.xticks(np.arange(min(x), max(x) + 1))  
plt.legend()

# Show the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Ques.5 Write a program in python to perform a multi-line plotted graph.

```python
import matplotlib.pyplot as plt

# functions for different lines
def f1(x):
    return 2 * x + 1

def f2(x):
    return x**2 - 3

def f3(x):
    return -x + 4


x = range(-5, 6)


y1 = [f1(i) for i in x]
y2 = [f2(i) for i in x]
y3 = [f3(i) for i in x]


plt.figure(figsize=(8, 6))
plt.plot(x, y1, label='f1(x) = 2x + 1')
plt.plot(x, y2, label='f2(x) = x^2 - 3')
plt.plot(x, y3, label='f3(x) = -x + 4')


plt.xlabel('x')
plt.ylabel('y')
plt.title('Multi-Line Plot')
plt.legend()


plt.grid(True)
plt.tight_layout()
plt.show()
```
